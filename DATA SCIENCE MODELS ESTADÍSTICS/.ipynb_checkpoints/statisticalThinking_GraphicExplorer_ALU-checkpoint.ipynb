## Graphical Exploratory Data Analysis

### HISTOGRAMA


Vamos a importar el dataset Iris un dataset generado por el botànico Edward Anderson donde analizaba tres espècies diferentes de la planta Iris y que da mucho juego a nivel estadìstico

```python
import pandas as pd

df_iris = pd.read_csv('iris.csv')

display(df_iris)
```

```python
#creamos una serie solo on la longitud de un tipo de concreto

versicolorPetalLenght = df_iris[df_iris['variety']=='Versicolor']['petal.length']
versicolorPetalLenght
```

```python
# import plotting modulo
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# añadimos el seabons por defecto

sns.set()

#num bins aconsejable

nbins = int(np.sqrt(len(versicolorPetalLenght)))

# mostramos istograma

plt.hist(versicolorPetalLenght, bins = nbins)

#añadimos etiquetas ejes

plt.xlabel('longitud petalo cm')
plt.ylabel('numero total')

plt.show()

```

### BEE SWARM PLOT (PANEL DE ABEJAS)

```python
#creamos panel de abejas
sns.swarmplot(x = 'variety', y = 'petal.length', data = df_iris)
plt.xlabel('variedad')
plt.ylabel('longitud petalo cm')

plt.show()
```

### ECDF(estimador de funcion de distribucion acumulada)

```python

def ecdf(data):
    #nmeor de datos
    n = len(data)
    #ordenamos datos 
    x = np.sort(data)
    # valores y
    y = np.arange(1,n+1) /n
    
    return x, y

x_vers , y_vers = ecdf(versicolorPetalLenght)
plt.plot(x_vers , y_vers, marker = '.', linestyle='none')

plt.xlabel('longitud de petalo')
plt.ylabel('ecdf')

plt.margins(0.02)
plt.show()
```

### BOXPLOT

```python
sns.boxplot(x = 'variety', y = 'petal.length', data = df_iris)
plt.xlabel('variety')
plt.ylabel('petal length cm')
```

### MEAN, MEDIAN, VARIANCE AND STD

```python
np.std(versicolorPetalLenght)
```

```python
import random

def mostrar_menu(opciones):
    # muestra los menos recorriendo el diccionario
    print('Seleccione una opción:')
    for clave in sorted(opciones):
        print(f' {clave}) {opciones[clave][0]}')

def leer_opcion(opciones):
    #le la opcion del usuario asta que no sea correcta pedira opcion
    while (a := input('Opción: ')) not in opciones:
        print('Opción incorrecta, vuelva a intentarlo.')
    return a

def is_entero():
    # comprobar que el input sea numero entero sino volvera a pedir
    while True:
        entrada = input("Escribe un numero entero: ")
        try:
            entrada = int(entrada)
            return entrada
        except ValueError:
            print ("La entrada es incorrecta: escribe un numero entero")
        
def ejecutar_opcion(opcion, opciones):
    #ejecutara la funcion asignada.
    opciones[opcion][1]()

def generar_menu(opciones, opcion_salida):
    #mientras la opcion dle user sea diferente a la de salida mostrara el menu, leera la opcion y la realizara
    opcion = None
    while opcion != opcion_salida:
        mostrar_menu(opciones)
        opcion = leer_opcion(opciones)
        ejecutar_opcion(opcion, opciones)
        print()

def menu_principal():
    #diccionario que contiene los datos del menu
    opciones = {
        '1': ('Adivina número', adivinaNumero),
        '2': ('Pedra - Paper - Tisores', papelTijera),
        '3': ('Penjat', penjat),
        '4': ('Salir', salir)
    }

    generar_menu(opciones, '4')
    
def adivinaNumero():
    # funciona del primer juego
    intentos = 3
    # generamos un numero entero del 1 al 10
    numeroR = random.randint(1, 10)
    print(" ¡Vamos a jugar Adivina el Numero! \n tienes 3 intentos para Adivinar \n solo sera del 1 al 10 ¡animo :D!")
    #mientras los intentos del usuario sean mayor que 0 
    while intentos > 0:
        print(f"intentos: {intentos} \n ¡RECUERDA! del 1 al 10 sino sera fallo :C")
        # se prueva que le nuemro sea entero
        numeroElegido = is_entero()
        # restamos 1 inteto 
        intentos -= 1
        # comprobamos que el entero este dentor del rango 1 y 10
        if numeroElegido < 1 or numeroElegido > 10:
            print(f"te abise >:|")
        else:
            if numeroElegido < numeroR:
                print("Frio frio, nuemero muy bajo :S")
            elif numeroElegido > numeroR:
                print("Calor calor, numero muy alto :´c")
            else:
                #si gana pase lo que pase salimos dle bucle
                print("¡BIEN HAS GANADO! :D")
                break
     # si no ganas te muestra un mensaje triste           
    if numeroElegido != numeroR:
        print("PERDISTE, que pena :C.")
        
        
def  comprobarGanador(opcion, numeroR, jugador, maquina):
    # comprbacion de quien es el ganador de la ronda
    if opcion == 1 and numeroR == 3:
        return True
    elif opcion == 2 and numeroR == 1:
        return True
    elif opcion == 3 and numeroR == 2:
        return True
    elif opcion == numeroR:
        return "empate"
    else:
        return False
        
                
def papelTijera():
    # juego numero 2, el usuario y la maquina tiene una puntuacion de 0
    print(" ¡Vamos a jugar Piedra Papel Tijera! \n ganara el que hacierte 3  \n ¡animo :D!")
    jugador = 0
    maquina = 0
    # jugara asta que alguien llege a 3
    while True:
        
        if jugador >= 3 or maquina >= 3:
            break
        # ranodm de 1 al 3 para la respuesta d la maquina
        numeroR = random.randint(1, 4)
        print( "1. Piedra \n 2. Papel \n 3.Tijera \n Elige una opcion: :D")
        #comprobamos que introduce un numero entero
        opcion = is_entero()
        # comprobamos que este entre el rango 1 y 3
        if opcion < 1 or opcion > 3:
            
            print(f"escribe 1 o 2 o 3 ¡animo :D!")
        else:
            
            print(f"has escogido : {eleccionJuegoDos(opcion)}")
            print(f"la maquina a escogido : {eleccionJuegoDos(numeroR)}")
            print("------------------")
            #almacemos el resultado del ganador
            winRonda = comprobarGanador(opcion, numeroR, jugador, maquina)
            
            if winRonda == True:
                print(f"Ganaste, de momento...")
                jugador +=1
                
            elif winRonda == "empate":
                print("EMPATE")
                
            else:
                print("perdiste, XD el azar...")
                maquina +=1
                
            print(f"total: \n JUGADOR : {jugador} \n MAQUINA :{maquina}")
            
    #muestra un mensaje alegre o triste si ganas               
    if(jugador == 3):
        print("MUY BIEN, LE GANASTE A UN PROGRAMA :D")
        
    else:
        print("PERDISTE, que pena :C.")
        
def addDatosColgado(contenido):
    #generamos los datos necesarios para jugar al ahorcado, las eleccion de la palabra, los intetos = longitud palabra *2
    # almacenamos la palabra, generamos una matriz llena de "_" que sera el tablero , y generamos una matriz para las letras erroneas
    randomEleccion = random.randint(0,len(contenido)-1)
    intentosTotal = len(contenido[randomEleccion]*2)
    palabra = contenido[randomEleccion].lower()
    tablero= ['_']*len(contenido[randomEleccion])
    return randomEleccion, intentosTotal, tablero, palabra, []

def TableroAhorcado(tablero,letrasError):
    # mostramos el table con un bucle si al no tiene mas char termina
    for x in tablero:
        print(x, end=" ")
        
    print(f"\n")
    # mostramos si la matriz tiene contenido las letras erroneas que no coinciden con las de la palabra
    if len(letrasError) > 0:
        print(f"LETRAS INCORRECTAS : ", *letrasError)
        print("----------------------")
        
def pedirLetra(tablero, letrasError):
    # pedimos una letra al usuario entre a-z siempre minuscula si no es letra volvera a pedir
    correcto = False      
    while not correcto:
        letra = input("introduce letra : (a-z): ").lower()
        correcto = 'a' <= letra <= 'z' and len(letra) == 1 
        if not correcto:
            print("Error, tiene que ser letra entre a y z")
        else:
            # se hace comprobacion que con "not in" que no encuentre  letra en las listas con el + las juntamos 
            correcto = letra not in tablero + letrasError
            if not correcto:
                print("LETRA REPETIDA")
    return letra

def acierto_Fallo(letraUser, palabra, tablero, letrasError):
    # si encuentra la letra en la palabra a hacertado sino añade la letra en la matriz de fallos
    if letraUser in palabra:
        print("Has acertado la letra :D!")
        actualizarTablero(letraUser,palabra,tablero)
        return True
    else:
        print("Has fallado, que pena :c")
        letrasError.append(letraUser)
        return False
    
def actualizarTablero(letraUser,palabra,tablero):
    #recorre la palabra en busca de la letra si la encuentra la sustituye por la "_" de tablero
    for x, letraPalabra in enumerate(palabra):
        if letraUser == letraPalabra:
            tablero[x] = letraUser
        
def palabraCompleta(tablero):
    # si no enucentra el sinbolo en la matriz es que le user a ganado :D
    return '_' not in tablero

def penjat():
    #juego numero 3 
    print(" ¡Vamos al Ahorcado! \n ganaras si completas la fase  \n ¡animo :D!")
    #leemos el fichero y almacenos todas las palabras en una matriz
    contenido = []
    with open('palabrasColgado.txt') as f:
        for line in f:
            contenido.append(line.strip())
    # datos con los que trabaja el codigo
    intentosUser  = 0
    randomEleccion,intentosTotal,tablero,palabra,letrasError= addDatosColgado(contenido)
    
    print(f"tienes un total de {intentosTotal} intents {len(letrasError)}")
    # mientras los intentos sean mas pequeños que los intentos totales jugara
    while intentosUser < intentosTotal:
        TableroAhorcado(tablero,letrasError)
        letraUser = pedirLetra(tablero, letrasError)
        #comprobamos que si no a hacertado le reste intento
        if acierto_Fallo(letraUser, palabra, tablero, letrasError) == False:
            intentosUser += 1
            print(f"llevas {intentosUser} de {intentosTotal} que tienes.")
        # si ganas sale directo del bucle
        if palabraCompleta(tablero):
            print("¡ HAS GANADO COLEGA.!")
            break
    else:
        print(f"PERDISTE, que pena :C.{contenido[randomEleccion]}")
    # mostramos otra vez el tablero para ver el resultado
    TableroAhorcado(tablero,letrasError)
    
def salir():
    #mensaje de fin de programa 
    print("finish")
    
def eleccionJuegoDos(opcion):
    #valida que opcion a escogido el usuario para jugar al piedra- papel- tijera
    if opcion == 1:
        return "Piedra"
    elif opcion == 2:
        return "Papel"
    else:
        return "Tijera"
    
#inicia el programa    
menu_principal()
```

El fichero «titanic.csv» contiene información sobre los pasajeros del Titanic. Escribir un programa
con los siguientes requisitos:

1. Generar un DataFrame con los datos del fichero.

```python
import pandas as pd
dataTitanic = pd.read_csv('titanic.csv')
```

2.Mostrar por pantalla las dimensiones del DataFrame, el número de datos que contiene, los
nombres de sus columnas, los tipos de datos de las columnas, las 10 primeras filas y las 10
últimas filas

```python
# numero de filas y columnas
print(dataTitanic.shape)
#todo lo demas
display(dataTitanic)
#el tipo de variable
print(dataTitanic.dtypes)
#coje los 10 primeros
print(dataTitanic[0:10])
#coje los 10 ultimos
print(dataTitanic.tail(10))
```

3. Mostrar por pantalla los datos del pasajero con identificador 148.


```python
pasajero = dataTitanic.query('PassengerId == 148')

print(pasajero)
```

4. Mostrar por pantalla las filas pares del DataFrame.


```python
dataTitanic.iloc[::2]
```

5. Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas
alfabéticamente.

```python
primeraClase = dataTitanic[dataTitanic['Pclass'] == 1]
ordenadoAlfa = primeraClase.sort_values(by = 'Name')
print(ordenadoAlfa)
```

6. Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.


```python
#el value_counts cuenta agrupa los valores y los cuenta
#el count cuenta todos los valores
porcentajeVivosDead = dataTitanic['Survived'].value_counts()/dataTitanic['Survived'].count() * 100
print(porcentajeVivosDead)
```

7. Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.


```python
#agrupamos las clases sociales luego de los que subrevicieron de cada clase los contamos y hacemos porcentaje
porcentajeXClase = dataTitanic.groupby(['Pclass'])['Survived'].value_counts() * 100

print(porcentajeXClase)
```

8. Eliminar del DataFrame los pasajeros con edad desconocida.


```python
dataTitanic.dropna(subset=['Age'])
```

9. Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.


```python
mujeres = dataTitanic.groupby(['Pclass','Sex'])['Age'].mean().unstack()['female']
print(mujeres)
```

10. Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.


```python
dataTitanic['MenorOrNot'] = dataTitanic['Age'] < 18
```

11. Mostrar por pantalla el porcentaje de menores y mayores de edad que sobrevivieron en cada
clase.

```python
menoresClass = dataTitanic.groupby(['Pclass', 'MenorOrNot'])['Survived'].value_counts(normalize = True) * 100
print(menoresClass)
```
